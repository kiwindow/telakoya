# 5. 複数シートの読み込み

## 1) 横方向の突合 (IDなどで突合)

- dataM.xlsxを読み込み、3つのシートを1つに突合する。
- まず、同一のIDを持つデータを1つにまとめるため、横方向(行方向)に突合する。➡ axis = 1

```python
# dataM.xlsxの読み込み (30sec)
dfM = pd.read_excel('/content/drive/MyDrive/Telakoya/tela3/data/dataM.xlsx', sheet_name = None)
```

```python
# データのタイプ
type(dfM)
```

```python
# データの長さ(要素数)
len(dfM)
```

```python
# 辞書形式のデータのキー
dfM.keys()
```

```python
# Sheet1の取り出し
df1 = dfM['Sheet1']
df1
```

```python
# Sheet2の取り出し
df2 = dfM['Sheet2']
df2
```

```python
# Sheet3の取り出し
df3 = dfM['Sheet3']
df3
```

```python
# タイプの確認
type(df1)
```

```python
# 横方向の突合
dfV = pd.concat([df1, df2, df3], axis = 1)
```

```python
# データの表示
dfV
```

```python
# 行と列の数
dfV.shape
```

## 2) 縦方向の突合 (説明変数で突合)

- dataL.xlsxというファイルを読み込み、3つのシートを1つに突合する。
- 同一の説明変数のデータが同一列になるよう、縦方向(列方向)に突合する。➡ axis = 0

```python
# dataL.xlsxの読み込み
dfL = pd.read_excel('/content/drive/MyDrive/Telakoya/tela3/data/dataL.xlsx', sheet_name = None)
```

```python
#　データのタイプ
type(dfL)
```

```python
# 辞書形式のデータのキー
dfL.keys()
```

```python
# Sheet1の取り出し
dfL1 = dfL["Sheet1"]
dfL1
```

```python
# Sheet2の取り出し
dfL2 = dfL["Sheet2"]
dfL2
```

```python
# Sheet3の取り出し
dfL3 = dfL["Sheet3"]
dfL3
```

```python
# 縦(column)方向の突合 (axis = 0は省略可)
dfH = pd.concat([dfL1, dfL2, dfL3], axis = 0)
```

```python
# データの表示
dfH
```

```python
# 行と列の数
dfH.shape
```
